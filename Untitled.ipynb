```python
#Importando Bibliotecas
import requests
import pandas
from bs4 import BeautifulSoup
import csv
```

## Gerando os links de todas as paginas de cada Gênero

```python
generos = ['action', 'adventure', 'animation', 'biography', 'comedy', 'crime', 'documentary', 'drama', 'family', 'fantasy', 'film', 'noir', 'history', 'horror', 'music', 'musical', 'mystery', 'romance' , 'sci-Fi', 'short Film', 'sport' 'superhero', 'thriller', 'war', 'western']
pages = []
           
for i in generos:
    url = 'https://www.imdb.com/search/title/?genres=action' + str(i) + '&title_type=feature&explore=genres&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=facfbd0c-6f3d-4c05-9348-22eebd58852e&pf_rd_r=BKXNX2HDBWJA129ZP7HB&pf_rd_s=center-6&pf_rd_t=15051&pf_rd_i=genre&ref_=ft_gnr_mvpop_1'              
    pages.append(url)
```

## Baixando o HTML do site 

```python
f = csv.writer(open('Desempenho de Filmes.csv', 'w'))
f.writerow(['Name', 'Ano', 'Gênero', 'Nota'])

for item in pages:
    page = requests.get(item)
    soup = BeautifulSoup(page.text, 'html.parser')
    
```

### Selecionando os items desejados no HTML (Nome, Ano, Gênero, Nota)

```python
last_links = soup.find(class_='lister-item-image')
last_links.decompose()
lista_lixo = soup.find_all(rel='nofollow')
lista_lixo.decompose()
    
lista_nome_filmes = soup.find(class_='lister-list')
nome_filmes = lista_nome_filmes.find_all('a')
ano_filme = lista_nome_filmes.find_all('span', 'lister-item-year')
```
